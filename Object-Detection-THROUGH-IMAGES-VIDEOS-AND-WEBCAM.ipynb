# video demo

```python
cap = cv2.VideoCapture("images/object-detection/small.mp4")

# Check if the webcam is opened correctly
if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Cannot open video source")

font_scale = 1  # Changed from 3 to 1
font = cv2.FONT_HERSHEY_PLAIN

while True:
    ret, frame = cap.read()
    
    if not ret:
        print("End of video or cannot read frame")
        break
    
    # Resize frame for better detection (optional but recommended)
    height, width = frame.shape[:2]
    if width > 640:
        scale = 640 / width
        frame = cv2.resize(frame, (640, int(height * scale)))
    
    ClassIndex, confidence, bbox = model.detect(frame, confThreshold=0.5)  # Fixed typo
    
    print(ClassIndex)
    
    if len(ClassIndex) != 0:
        for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
            if ClassInd <= 80:
                cv2.rectangle(frame, boxes, (255, 0, 0), 2)
                
                # Fixed: Use ClassInd-1 for 0-indexed arrays or just ClassInd if already correct
                label = f"{classLabels[ClassInd-1]}: {conf:.2f}"  # Show confidence too
                
                cv2.putText(frame, label, (boxes[0]+10, boxes[1]+40), 
                           font, fontScale=font_scale, color=(0, 255, 0), thickness=2)
    
    cv2.imshow('Object Detection Tutorial', frame)
    
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

# webcam-demo

```python
cap = cv2.VideoCapture(1)

# Check if the webcam is opened correctly
if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Cannot open webcam")

# Set webcam resolution for better performance
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

font_scale = 1  # Changed from 3
font = cv2.FONT_HERSHEY_PLAIN

while True:
    ret, frame = cap.read()
    
    if not ret:
        print("Failed to grab frame")
        break
    
    # Detect objects
    ClassIndex, confidence, bbox = model.detect(frame, confThreshold=0.45)  # Fixed typo & lowered threshold
    
    print(ClassIndex)
    
    if len(ClassIndex) != 0:
        for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
            if ClassInd <= 80:
                # Draw rectangle
                cv2.rectangle(frame, boxes, (255, 0, 0), 2)
                
                # Fixed class index and added confidence display
                class_name = classLabels[ClassInd - 1]  # Fixed from ClassInd-10
                label = f"{class_name}: {conf:.2f}"
                
                cv2.putText(frame, label, (boxes[0] + 10, boxes[1] + 40), 
                           font, fontScale=font_scale, color=(0, 255, 0), thickness=2)
    
    cv2.imshow('Object Detection Tutorial', frame)
    
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
